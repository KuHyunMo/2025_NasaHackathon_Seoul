환경 설정 및 데이터 불러오기

```python
# 필요한 라이브러리 설치
# !pip install astropy tensorflow # 이미 설치했다면 주석 처리

import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt


# FITS 파일 열기 및 데이터 확인 함수 (수정 없음)
def load_fits_image(file_path):
    try:
        with fits.open(file_path) as hdul:
            print(f"Inspecting FITS file: {file_path}")
            hdul.info()

            if len(hdul) > 1 and hdul[1].data is not None:
                image_data = hdul[1].data.astype(np.float32)

                image_data = np.nan_to_num(image_data)

                min_val, max_val = np.min(image_data), np.max(image_data)

                if min_val == max_val:
                    print("Warning: Image data has uniform values.")
                    return np.zeros_like(image_data)

                image_data = (image_data - min_val) / (max_val - min_val)
                return image_data
            else:
                # SCI 데이터가 없는 경우 첫 번째 HDU를 시도
                print("Warning: Second HDU ('SCI') not found or empty, trying the first HDU.")
                if hdul[0].data is not None:
                    image_data = hdul[0].data.astype(np.float32)
                    image_data = np.nan_to_num(image_data)
                    min_val, max_val = np.min(image_data), np.max(image_data)
                    if min_val == max_val:
                        return np.zeros_like(image_data)
                    image_data = (image_data - min_val) / (max_val - min_val)
                    return image_data
                else:
                    print("Error: No valid image data found in any HDU.")
                    return None

    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


# 예시 FITS 파일 경로
fits_file = 'data/hlsp_heritage_hst_acs-wfc_m51_f555w_v1_drz_sci.fits'
image_2d = load_fits_image(fits_file)

# 이미지가 잘 불러와졌는지 시각화하여 확인
if image_2d is not None:
    # ## <<< 변경된 부분 시작 ###
    # 로그 스케일링 적용
    # np.log1p는 log(1+x)를 계산하여 0 값에 대한 에러를 방지합니다.
    log_scaled_image = np.log1p(image_2d)

    # 원본 이미지와 로그 스케일링된 이미지를 함께 출력하여 비교
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image_2d, cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title('Log Scaled Image')
    plt.imshow(log_scaled_image, cmap='gray')

    plt.show()
    # ## <<< 변경된 부분 끝 ###

    print("이미지 크기:", image_2d.shape)
    print("데이터 타입:", image_2d.dtype)
```
